```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("StudentsPerformance.csv")
data
```

```python
data.isnull().sum()
```

```python
sns.set(style='whitegrid')
ax=sns.barplot(x=data['gender'].value_counts().index,
              y=data['gender'].value_counts().values,palette="OrRd_r",
              hue=['Female','Male'])
plt.legend(loc=8)
plt.xlabel('Gender')
plt.ylabel('Frequency')
plt.title('Show of Gender Bar Plot')
plt.show()
```

```python
plt.figure(figsize=(10,7))
sns.barplot(x = "parental level of education",
           y="writing score",
           hue='gender',
            data=data
           )
plt.show()
```

```python
label=data['race/ethnicity'].value_counts().index
colors=['blue','red','yellow','green','brown']
explode=[0,0,0.1,0,0]
values=data['race/ethnicity'].value_counts().values
plt.figure(figsize=(7,7))
plt.pie(values,explode=explode, labels=label,colors=colors,autopct='%1.1f%%')
plt.title('race/ethnicity Analysis',color='black',fontsize=10)
plt.show()
```
